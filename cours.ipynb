# Introduction

## Avant-propos : texte brut, texte enrichi

Un fichier de texte peut contenir des indications de mise en forme

- `texte brut` (ou `plain text`) = un texte dépourvu d'information de mise en forme, qui se limite à l'affichage de la traduction des données binaires en caractères alphanumériques et symboles

- `texte enrichi` (ou `rich text`) = un texte qui comporte des éléments de mise en forme, qu'ils soient ou non affichés dans l'interface graphique

## Éditeur de texte, traitement de texte = deux paradigmes

- `WYSIWYG` (*What You See Is What You Get*) = un `traitement de texte` qui présente un texte enrichi avec sa mise en forme, de manière simultanée

- `WYSIWYM` (*What You See Is What You Mean*) = un `éditeur de texte` qui présente le texte et ses informations de mise en forme sans les interpréter

## Histoire et principes

🇱 1978 — Donald Knuth initiative du langage TEX (/tɛx/ : tau, epsilon, chi (τεχ)) pour éditer son œuvre majeure *The Art of Computer Programming*

🇱 un langage et un système de composition de documents, grande précision typographique

🇱 logiciel libre et gratuit

🇱 1984 — LaTeX : simplification par Leslie Lamport

🇱 utilisé dans le domaine des sciences car on peut facilement écrire des formules mathématiques

## Il faut distinguer…

🇱 TeX (le moteur de base) ≠ LaTeX (ensemble d'outils pour la composition)

🇱 LaTeX = un langage + un moteur de composition + des packages qui étendent les fonctionnalités de TeX + des polices…

🇱 LaTeX désigne donc un `ensemble logiciel complet` 

🇱 Les extensions de TeX, de nouveaux moteurs pour palier les limites du programme historique : pdfTeX xeTeX et luaTeX

🇱 xeLaTeX, pdfLaTeX, luaLaTeX = une surcouche pour utiliser les macros LaTeX

🇱 pas des balises sémantiques mais des commandes `\chapter{Titre de chapitre}`


## Taper en LaTeX

- le nombre d'espaces ne change rien
- un retour à la ligne = un espace
- un tilde (`~`) = espace insécable
- deux retours à la ligne ou plus = nouveau paragraphe
- `--` = demi-cadratin, `---` = cadratin
- pour taper des points de suspensions: `\dots`
- tout ce qui suit le caractère `%` est un commentaire
- attention, certains caractères sont utilisés par LaTeX: `\ ~ ^ { } % _ & $ #`. Pour insérer ces caractères dans un texte, il faut taper: `\textbackslash  \textasciitilde \textasciicircum \{ \} \% \_ \& \$ \#`

## Les environnements

- `environnement` = portion de document qui a une signification spécifique et subit un traitement spécifique. L'environnement accepte le saut de paragraphe. On peut utiliser des commandes au sein d'environnements.
- syntaxe d'un environnement = `\begin{nom_de_l_environnement} \end{nom_de_l_environnemen}`. Tout ce qui se trouve entre ces deux balises est à l'intérieur de l'environnement.

*exemple : l'environnement `abstract` de la classe `article` :*

## Les listes – 👩🏻‍💻 exercice 4

LaTeX propose trois environnement :

`itemize` — liste à puces

```
latex\begin{itemize}
  \item Premier élément
  \item Deuxième élément
\end{itemize}
```

`enumerate` — liste numérotée

```
latex\begin{enumerate}
  \item Première étape
  \item Deuxième étape
\end{enumerate}
```

`description` — liste de définitions. On place entre crochets après `\item` 
le terme à mettre en gras

```
latex\begin{description}
  \item[Terme] Définition du terme.
  \item[Autre mot] Sa description ici.
\end{description}
```

Les listes peuvent être imbriquées (jusqu'à 4 niveaux). Le package `enumitem` permet de personnaliser la mise en forme (espacement, style des puces, etc.).

## Les citations 

Les `guillemets` 

```
``Texte entre guillemets anglais''   % "Texte entre guillemets anglais"
\og Texte entre guillemets français \fg{}  % « Texte... » (avec package babel/frenchb)
\enquote{Cette commande permet de faire des \enquote{citations imbriquées}} %  Cette commande permet de faire des “citations imbriquées”. » (avec package csquote)
```

`quote` — pour une citation courte, sans indentation de première ligne

```
\begin{quote}
  Ceci est une courte citation.
\end{quote}
```

*NB* : Avec le package csquote, on peut aussi mettre les points de suspension entre crochets (`\textelp{}`) et insérer des commentaires entre crochets (`\textins{commentaire}`). 

`quotation` — pour une citation longue sur plusieurs paragraphes. Ajoute une indentation de première ligne à chaque paragraphe.

```
\begin{quotation}
  Premier paragraphe de la citation.

  Deuxième paragraphe, avec indentation automatique.
\end{quotation}
```

`verse` — pour la poésie. On passe au vers suivant avec `\\`, et on saute une ligne pour changer de strophe. Si un vers est trop long, la suite est indentée pour ne pas la confondre avec un nouveau vers. Pour rajouter une identation sur certains vers on rajoute `\>`

```
\begin{verse}
  Premier vers, \\
  Deuxième vers, \\
  Troisième vers.

  Début d'une nouvelle strophe, \\
  Et sa suite ici.
\end{verse}
```

## Références croisées – 👩🏻‍💻 exercice 6

Référencer une section

```
latex\section{Mon titre}
\label{sec:montitle}   % juste après \section

Voir la section~\ref{sec:montitle}.     % → "section 2"
Voir~\nameref{sec:montitle}.            % → "Mon titre" (avec hyperref)
Voir~\cref{sec:montitle}.              % → "section 2" (avec cleveref)
```

Référencer une figure

```
latex\begin{figure}
  \centering
  \includegraphics{image.png}
  \caption{Légende.}
  \label{fig:monimage}   % ⚠️ toujours APRÈS \caption
\end{figure}

La figure~\ref{fig:monimage} montre…   % → "La figure 1.5"
```

Référencer une équation

```
latex\begin{equation}
  E = mc^{2}
  \label{eq:einstein}
\end{equation}

L'équation~\eqref{eq:einstein}…   % → "(1)" avec parenthèses
```

**À retenir :**

- Le tilde `~` avant `\ref` pour une espace insécable
- Compiler deux fois pour résoudre les références
- Utiliser `\eqref` plutôt que `\ref` pour les équations (parenthèses)
- Voire la documentation du package `cleveref` (chargé en dernier) pour personnaliser

# Structurer des données bibliographiques

## Données vs. style biblio

1. La référence = donnée

Quand on est dans des pratiques de recherche et qu'on veut présenter ses sources, il y a deux actes différents : la citation à l'intérieur du texte et le fait de créer une bibliographie, qui constitue la liste des références à qu'on a utilisées à la fin du document.

Les références bibliographiques = des données

```
┌─────────────────────────────────────┐
│  FICHE CATALOGUE 📇                 │
├─────────────────────────────────────┤
│ Cote: 801.95 GEN                    │
│ Auteur: GENETTE, Gérard             │
│ Titre: Figures III                  │
│ Éditeur: Seuil                      │
│ Année: 1972                         │
│ Pages: 285 p.                       │
│ Collection: Poétique                │
│ ISBN: 2-02-002351-5                 │
└─────────────────────────────────────┘
```

Si ces données sont bien structurées, on peut les `réutiliser`, les `formater` (présenter) de plusieurs manière, et les `exposer`.

## Données vs. style biblio

2. Le style bibliographique = la présentation des données 

Les revues scientifique, les maisons d'édition ou les universités imposent des contraintes dans la présentation des références bibligraphiques (citation ou bibliographie). 

Un style bibliographique = un ensemble de règles pour mettre en forme la référence bibliographique. 

**Style APA (7e édition)**<br/>
Genette, G. (1972). *Figures III*. Éditions du Seuil.

**Style MLA (9e édition)**<br/>
Genette, Gérard. *Figures III*. Éditions du Seuil, 1972.

**Style Chicago (english)**<br/>
Genette, Gérard. *Figures III*. Paris: Éditions du Seuil, 1972.

Il faut donc produire une balisage pour identifier tous les éléments d'une référence bibliographique.

```
DONNÉES (contenu)          STYLE (présentation)
     ↓                            ↓
[Base de données]    →    [Règles de formatage]
                                   ↓
                          [Référence formatée]
```

## Comment ça se passe en latex 

### Principe 

1. Un fichier au format BibTeX, dont l'extension est `.bib`, et qui contient toutes les métadonnées de vos références bibliographique

2. À chaque entrée est associée une clé qui l'identifie

3. Puis chaque entrée fait une description structurée de la référence : nature, titre, date, auteur… Le nom des champs est standardisé.

## Revenons à nos en-têtes et pieds de page

Pour afficher les contenus dynamiquement, LaTeX les fait passer par plusieurs commandes qui s'appellent les unes les autres. 

```
\chapter  →  alimente \chaptermark  →  appelle \markboth  →  alimente \leftmark et \rightmark
\section  →  alimente \sectionmark  →  appelle \markright →  alimente \rightmark
```

En pratique on ne touche jamais à `\markboth` directement. On redéfinit `\chaptermark` ou `\sectionmark` avec `\renewcommand`. On s'insère dans cette chaîne au bon endroit, sans toucher à `\chapter` lui-même.

## Revenons aux en-têtes et pieds de page

Par défaut `\chaptermark` est défini ainsi :

```latex
\renewcommand{\chaptermark}[1]{
    \markboth{\MakeUppercase{\chaptername\ \thechapter.\ #1}}{}
}
```
où : 

- `\thechapter` donne le numéro du chapitre
- `#1` est le titre du chapitre
- `\markboth{}{}` prend deux arguments : gauche et droite

## Écriture vs. saisie de données

> \[L\]e formalisme de la tradition imprimée n’est pas de la même nature que celui de la pratique numérique car les conventions suivies dans les éditions imprimées sont en fait le produit de changements transmis de génération en génération durant plus de trois siècles. La maîtrise de ces conventions s’apparente donc à celle d’une langue pourvue de signes, de termes techniques, de règles et d’ornements de style où le latin occupe la première place. Connaître cette langue, c’est aussi la connaître dans le temps et être en mesure de la suivre à une époque où elle n’était pas codifiée comme elle l’est aujourd’hui par les grandes maisons d’édition. La lecture méthodique de cette langue requiert certes un entraînement particulier, mais elle repose sur l’intelligence humaine et sa faculté d’interpréter naturellement les termes, les signes et l’économie de leur disposition. Et pour la même raison, cette lecture reste inaccessible à une machine. Autrement dit, tandis que les notes critiques des éditions imprimées sont écrites dans une langue suivant une grammaire qui lui est propre, celles des éditions numériques sont saisies dans un code structuré formé d’éléments subordonnés et d’attributs, chacun destiné à recevoir un contenu qui lui est propre. (Alessi 2023)

## Saisir l'apparat 👩🏻‍💻

Le texte de l'édition se place dans un environnement ekdosis. À chaque endroit où il y a une variante, on insère la commande `\app`, qui contient le lemme retenu (`\lem`) et les variantes rejetées (`\rdg`, de l'anglais reading). 

```latex
\begin{ekdosis}
Sunt vestrum aliquam multi qui L. Pisonem
\app{
  \lem[wit={S, D}]{cognorint}
  \rdg[wit={ψ}]{gnorunt}
  \rdg[wit={R}]{gonrit}
  \rdg[wit={π}]{cognoverint}},
huius L. Pisonis, qui praetor fuit, patrem.
\end{ekdosis}
```
Dans cet exemple, la leçon retenue est cognorint, attestée par les témoins S et D. Les trois autres `\rdg` présentent les variantes des témoins ψ, R et π. ekdosis se charge lui-même de composer l'apparat en bas de page dans la mise en page traditionnelle.

## Anatomie détaillée de la commande `\app` 👩🏻‍💻

```latex
\app{
  \lem[wit={P,R,a,c}]{formosam}
  \rdg[source={Asper}, alt={---onsam}]{formonsam}
  \note{(\emph{Append. Seruiana})}
}
```

`\lem` pour `Lemma` : la leçon retenue dans le texte. L'argument `wit=` ou `source=` liste les manuscrits ou les sources qui la portent.

`\rdg` pour `Reading` : une variante rejetée. `source=` pour sa source, `wit=` pour le ms. qui la présente, `alt=` pour la forme abrégée à afficher (mais il faut la forme complète pour la sortie TEI).
`\note` pour une note libre dans l'apparat (référence bibliographique, explications).



## Aligner le texte avec une traduction 👩🏻‍💻

Pour présenter le texte et sa traduction sur la même page en deux colonnes, remplacez l'environnement `ekdosis` par un environnement alignment contenant deux sous-environnements `edition` et `translation` (modifiable dans le préambule) :

```latex
\begin{alignment}[tcols=2, lcols=2]
\begin{edition}
  … texte latin avec \app{} …
\end{edition}
\begin{translation}
  … traduction française …
\end{translation}
\end{alignment}
```

L'option `tcols=2` indique deux colonnes au total, et `lcols=2` donne le nombre de colonne, parmi le total, qui s'imprime sur la page de gauche.
